```python
def refine_csv(filename):
    data = pd.read_csv(filename)
    d1 = data[data!="ingc-team@amazon.com"]
    d2 = d1[pd.notnull(d1.Email)]
    return d2
```

```python
new_data = pd.read_csv("sharp.csv")
new_data
```

```python
def get_emails():
    a = []
    for email in new_data.Email:
        if email not in a :
            a.append(email)
    return a
emails = get_emails()
```

```python
nn = []
for i,j in new_data.iterrows():
    if j.Email in emails:
        nn.append(j)
        emails.remove(j.Email)
```

```python
pd.DataFrame(nn).to_csv("Final_lis.csv")
```
